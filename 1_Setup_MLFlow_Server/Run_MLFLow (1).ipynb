```python
import os
import subprocess
from urllib.request import pathname2url
from pyngrok import ngrok

# Force ngrok to use IPv4 (fixes ngrok connection refused issue on some systems)
os.environ["NGROK_FORCE_IPV4"] = "true"

# Convert Windows paths to proper file:// URIs
def to_file_uri(path):
    return "file:///" + pathname2url(os.path.abspath(path)).replace("\\", "/")

# Configuration
mlflow_host = "127.0.0.1"         # Bind MLflow to localhost
mlflow_port = "5000"
backend_store = "./mlruns"
artifact_store = "./artifacts"

# Ensure directories exist
os.makedirs(backend_store, exist_ok=True)
os.makedirs(artifact_store, exist_ok=True)

# Convert local paths to file URIs for MLflow
backend_store_uri = to_file_uri(backend_store)
artifact_store_uri = to_file_uri(artifact_store)

# Build MLflow server command
command = [
    "mlflow", "server",
    "--host", mlflow_host,
    "--port", mlflow_port,
    "--backend-store-uri", backend_store_uri,
    "--default-artifact-root", artifact_store_uri
]

print(f"üöÄ Launching MLflow server at http://{mlflow_host}:{mlflow_port}")
print(f"üìÅ Backend URI: {backend_store_uri}")
print(f"üì¶ Artifact URI: {artifact_store_uri}\n")

# Step 1: Open ngrok tunnel to MLflow port
public_url = ngrok.connect(addr=int(mlflow_port), proto="http")
print(f"üåç Public MLflow URL: {public_url}")

# Step 2: Run MLflow server in a subprocess (blocking call)
try:
    subprocess.run(command)
except KeyboardInterrupt:
    print("\n‚ùå MLflow server stopped manually.")
finally:
    print("üîí Closing ngrok tunnel...")
    ngrok.disconnect(public_url)
    ngrok.kill()
```
